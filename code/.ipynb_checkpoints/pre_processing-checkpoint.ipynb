# 数据预处理

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# 读取每个 Excel 文件
df1 = pd.read_excel('../data/RESSET_DRESSTK_2016_2020_1.xlsx')
df2 = pd.read_excel('../data/RESSET_DRESSTK_2011_2015_1.xlsx')
df3 = pd.read_excel('../data/RESSET_DRESSTK_2021__1.xlsx')

# 合并数据框
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
combined_df.head()
```

```python
# 打印 combined_df 的基本信息
combined_df.info()
```

```python
total_rows = len(combined_df)
print(f"Total number of rows: {total_rows}")
```

## 数据清洗

```python
# 删除指定列
columns_to_drop = [
    '股票拆细比例(股)_Stksprate', '缩股比例(股)_Shrcmprsrate', '增发价格(元)_Snipr', 
    '增发新股数(股)_Snivol', '对价派现金额(元)_Pcash', '对价送股比例_Pshrrate', 
    '对价送股比例_公司送股_Pshrrate_C', '对价送股比例_非流通股东送股_Pshrrate_S', 
    '对价转增比例_Pcapissurate','除权除息日_Exdt','股数变动日_Capchgdt','现金红利(元)_Dividend',
    '送股比例(股)_Stkdrate','转增比例(股)_Capissurate','配股比例(股)_Rigoffrate','配股价格(元)_Rigoffpr',
    '实际配股数(股)_Arigoffshr'
]

# 删除列并更新数据框
combined_df = combined_df.drop(columns=columns_to_drop)

# 显示删除后的数据框信息
combined_df.info()
```

## 处理缺失值

```python
# 对于价格相关字段：使用相邻日期的值填充（如前一天的收盘价）或使用插值方法
combined_df['前收盘价_PrevClPr'] = combined_df['前收盘价_PrevClPr'].fillna(method='ffill')
combined_df['开盘价_Oppr'] = combined_df['开盘价_Oppr'].fillna(method='ffill')
combined_df['最高价_Hipr'] = combined_df['最高价_Hipr'].fillna(method='ffill')
combined_df['最低价_Lopr'] = combined_df['最低价_Lopr'].fillna(method='ffill')
combined_df['收盘价_Clpr'] = combined_df['收盘价_Clpr'].fillna(method='ffill')
combined_df['复权价1(元)_AdjClpr1'] = combined_df['复权价1(元)_AdjClpr1'].fillna(method='ffill')
combined_df['复权价2(元)_AdjClpr2'] = combined_df['复权价2(元)_AdjClpr2'].fillna(method='ffill')
```

```python
combined_df.info()
```

```python
# 对于没有成交量或成交金额的记录，填充为0表示当天没有交易
combined_df['成交量_Trdvol'] = combined_df['成交量_Trdvol'].fillna(0)
combined_df['成交金额_Trdsum'] = combined_df['成交金额_Trdsum'].fillna(0)
```

```python
# 日振幅、流通股日换手率和总股数日换手率使用等使用前向或后向填充，保持换手率的时间连续性
combined_df['日振幅(%)_Dampltd'] = combined_df['日振幅(%)_Dampltd'].fillna(method='ffill')
combined_df['总股数日换手率(%)_DFulTurnR'] = combined_df['总股数日换手率(%)_DFulTurnR'].fillna(method='ffill')
combined_df['流通股日换手率(%)_DTrdTurnR'] = combined_df['流通股日换手率(%)_DTrdTurnR'].fillna(method='ffill')
```

```python
# 公司国有股和法人股数据缺失用0填充
combined_df['公司国有股_Comstateshr'].fillna(0, inplace=True)
combined_df['公司法人股_Comlpshr'].fillna(0, inplace=True)
```

```python
# 考虑收益率和资本收益率、累积股价调整乘子等的数据点在时间序列中有一定的趋势，使用插值法来填充缺失值
combined_df['累积股价调整乘子_Mcfacpr'] = combined_df['累积股价调整乘子_Mcfacpr'].interpolate(method='linear')
combined_df['日收益率_Dret'] = combined_df['日收益率_Dret'].interpolate(method='linear')
combined_df['日资本收益率_Daret'] = combined_df['日资本收益率_Daret'].interpolate(method='linear')
combined_df['等权平均市场日收益率_Dreteq'] = combined_df['等权平均市场日收益率_Dreteq'].interpolate(method='linear')
combined_df['调整后每股净资产(元/股)_NAPSadj'].interpolate(method='linear', inplace=True)
columns_to_interpolate = [
    '流通市值加权平均市场日收益率_Drettmv',
    '总市值加权平均市场日收益率_Dretmc',
    '等权平均市场日资本收益率_Dareteq',
    '流通市值加权平均市场日资本收益_Darettmv',
    '总市值加权平均日资本收益_Daretmc',
    '日无风险收益率_DRfRet'
]

for col in columns_to_interpolate:
    combined_df[col] = combined_df[col].interpolate(method='linear')
```

```python
# 汇率在短时间内的波动通常较小，使用前向或后向填充方法，保持数据的连续性
combined_df['汇率_Ex'] = combined_df['汇率_Ex'].fillna(method='ffill').fillna(method='bfill')
```

```python
# 由于佣金和印花税通常稳定，使用均值填充来处理。
mean_coma = combined_df['A股佣金_Coma'].mean()
combined_df['A股佣金_Coma'] = combined_df['A股佣金_Coma'].fillna(mean_coma)
combined_df['市盈率_PE'] = combined_df.groupby('证监会行业大类代码_Csrciccd2')['市盈率_PE'].transform(lambda x: x.fillna(x.mean()))
combined_df['市净率_PB'] = combined_df['市净率_PB'].fillna(combined_df['市净率_PB'].mean())
columns_to_fill_mean = [
    'B股佣金_Comb',
    'A股印花税_Stamptaxa',
    'B股印花税_Stamptaxb',
    '市现率_PCF',
    '市销率_PS'
]
for col in columns_to_fill_mean:
    combined_df[col] = combined_df.groupby('证监会行业大类代码_Csrciccd2')[col].transform(lambda x: x.fillna(x.mean()))
```

```python
combined_df.to_excel('combined_data.xlsx', index=False)
```

## 去除重复值

```python
combined_df = pd.read_excel('combined_data.xlsx')
combined_df.drop_duplicates(inplace=True)
```

## 数据类型转换

```python
# 确保日期列是datetime类型
combined_df['日期_Date'] = pd.to_datetime(combined_df['日期_Date'])
# 将分类变量转换为category类型
combined_df['上市状态_Listedstate'] = combined_df['上市状态_Listedstate'].astype('category')
```

## 异常值处理

```python
# 例：根据IQR过滤掉异常值
Q1 = combined_df['收盘价_Clpr'].quantile(0.25)
Q3 = combined_df['收盘价_Clpr'].quantile(0.75)
IQR = Q3 - Q1
combined_df = combined_df[(combined_df['收盘价_Clpr'] >= (Q1 - 1.5 * IQR)) & 
                           (combined_df['收盘价_Clpr'] <= (Q3 + 1.5 * IQR))]
```

```python
# 绘制箱线图检查收益率和收盘价等关键指标是否仍存在异常值import matplotlib.pyplot as plt
from matplotlib import font_manager
import seaborn as sns
import matplotlib.pyplot as plt

# 设置中文字体
font_path = 'simhei.ttf'  # 字体文件路径，确保字体文件在同目录下
font_prop = font_manager.FontProperties(fname=font_path)

# 选择要绘制的列
price_columns = ['开盘价_Oppr', '最高价_Hipr', '最低价_Lopr', '收盘价_Clpr']
return_columns = ['日收益率_Dret', '日资本收益率_Daret', 
                  '等权平均市场日收益率_Dreteq', '流通市值加权平均市场日收益率_Drettmv',
                  '总市值加权平均市场日收益率_Dretmc', '等权平均市场日资本收益率_Dareteq', '日无风险收益率_DRfRet']

# 创建图形窗口，并设置大小
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# 绘制价格箱线图
sns.boxplot(data=combined_df[price_columns], ax=axes[0])
axes[0].set_title('价格指标的箱线图', fontproperties=font_prop)
axes[0].tick_params(axis='x', rotation=45)
for label in axes[0].get_xticklabels():
    label.set_fontproperties(font_prop)  # 设置 x 轴标签字体属性

# 绘制收益率箱线图
sns.boxplot(data=combined_df[return_columns], ax=axes[1])
axes[1].set_title('收益率指标的箱线图', fontproperties=font_prop)
axes[1].tick_params(axis='x', rotation=45)
for label in axes[1].get_xticklabels():
    label.set_fontproperties(font_prop)  # 设置 x 轴标签字体属性

# 调整布局
plt.tight_layout()
plt.show()
```

```python
# 对收益率数据进行对数转换，以减少极端值的影响，并使数据更加平滑
import numpy as np
import pandas as pd

# 读取数据
combined_df = pd.read_excel('combined_data.xlsx')

# 指定需要对数转换的收益率字段
return_columns = ['日收益率_Dret', '日资本收益率_Daret', 
                  '等权平均市场日收益率_Dreteq', '流通市值加权平均市场日收益率_Drettmv',
                  '总市值加权平均市场日收益率_Dretmc', '等权平均市场日资本收益率_Dareteq', '日无风险收益率_DRfRet']

# 对收益率字段进行对数转换（加1以避免负值问题）
for col in return_columns:
    combined_df[col] = np.log1p(combined_df[col])

# 检查转换结果
combined_df[return_columns].head()
```

```python
columns_to_check = [
    '流通市值加权平均市场日收益率_Drettmv',
    '总市值加权平均市场日收益率_Dretmc',
    '等权平均市场日资本收益率_Dareteq',
    '流通市值加权平均市场日资本收益_Darettmv',
    '总市值加权平均日资本收益_Daretmc',
    '日无风险收益率_DRfRet',
    '累积股价调整乘子_Mcfacpr',
    '日资本收益率_Daret',
    '等权平均市场日收益率_Dreteq',
    '调整后每股净资产(元/股)_NAPSadj'
]

# 计算每行中指定字段的非0值个数
non_zero_counts = (combined_df[columns_to_check] != 0).sum(axis=1)

# 找出非0值小于等于2的行数
count = (non_zero_counts <= 2).sum()

print(f"非0值小于等于2的行数: {count}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据
file_path = 'workdata_log.xlsx'
data = pd.read_excel(file_path)

# 计算每行中 0 值的个数
zero_counts = (data == 0).sum(axis=1)

# 可视化 0 值数量从 44 到 0 的行数变化折线图
zero_counts_distribution = [(zero_counts >= i).sum() for i in range(44, -1, -1)]
plt.figure(figsize=(10, 6))
plt.plot(range(44, -1, -1), zero_counts_distribution, marker='o')
plt.xlabel('Number of Zero Values per Row (Descending)')
plt.ylabel('Number of Rows')
plt.title('Change in Number of Rows with Zero Values from 44 to 0')
plt.grid(True)
plt.show()
```
